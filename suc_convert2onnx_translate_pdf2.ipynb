```python
import torch
from transformers import AutoTokenizer, pipeline
from optimum.onnxruntime import ORTModelForSeq2SeqLM
import os
import shutil
import time # لاستيراد مكتبة الوقت لحساب المدة

# --- Define Paths ---
parent_dir = "/content/opus-mt-en-ar"
onnx_model_path = os.path.join(parent_dir, "onnx")
hf_model_name = "Helsinki-NLP/opus-mt-en-ar" # Keep this for loading the tokenizer later

# --- !!! قم بتحديد مسار الملف المدخل والمخرج هنا !!! ---
input_file_path = "/content/1.txt"  # استبدل هذا بمسار ملف الإدخال الفعلي
output_file_path = "/content/output_arabic.txt" # استبدل هذا بالمسار المرغوب لملف الإخراج

# --- (اختياري) إنشاء ملف إدخال نموذجي إذا لم يكن موجودًا ---
if not os.path.exists(input_file_path):
    print(f"ملف الإدخال '{input_file_path}' غير موجود. سيتم إنشاء ملف نموذجي.")
    with open(input_file_path, 'w', encoding='utf-8') as f:
        f.write("Hello, world!\n")
        f.write("This is an example sentence.\n")
        f.write("Translate this text file line by line.\n")
        f.write("ONNX Runtime provides faster inference.\n")

# --- Ensure ONNX directory exists ---
os.makedirs(onnx_model_path, exist_ok=True)

# --- Copy necessary config/tokenizer files ---
print(f"Copying config/tokenizer files from {parent_dir} to {onnx_model_path}")
files_to_copy = [
    "config.json",
    "generation_config.json",
    "source.spm",
    "target.spm",
    "tokenizer_config.json",
    "vocab.json",
    "special_tokens_map.json"
]
copied_count = 0
for filename in files_to_copy:
    src_file = os.path.join(parent_dir, filename)
    dst_file = os.path.join(onnx_model_path, filename)
    if os.path.exists(src_file):
        if not os.path.exists(dst_file): # نسخ فقط إذا لم يكن موجوداً بالفعل
            try:
                shutil.copy2(src_file, dst_file)
                print(f"  Copied: {filename}")
                copied_count += 1
            except Exception as e:
                print(f"  Error copying {filename}: {e}")
        else:
            # print(f"  Skipping copy (already exists): {filename}") # يمكنك إلغاء التعليق إذا أردت رؤية هذا
             pass # الملف موجود بالفعل، لا داعي للنسخ أو الطباعة
    else:
        print(f"  Warning: Source file not found, skipping: {src_file}")

# --- Check if essential config file exists after attempting copy ---
config_present = os.path.exists(os.path.join(onnx_model_path, "config.json"))

if not config_present:
     print("\nError: Configuration file 'config.json' is missing in the ONNX directory.")
     print("Please ensure the necessary files were downloaded or copied correctly.")
     # Exit or handle error appropriately
else:
    # --- Verification Step ---
    print(f"\nChecking files in: {onnx_model_path}")
    if os.path.exists(onnx_model_path):
        # طباعة عدد قليل من الملفات لتجنب إخراج طويل جداً
        all_files = os.listdir(onnx_model_path)
        print(all_files[:10], "..." if len(all_files) > 10 else "")
    else:
        print("ONNX model directory not found!")

    # --- Load the ONNX Model and Tokenizer ---
    try:
        print("\nAttempting to load model...")
        start_load_time = time.time()
        model = ORTModelForSeq2SeqLM.from_pretrained(
            onnx_model_path,
            use_cache=True # استخدام الكاش لأننا قمنا بتنزيل ملف decoder_with_past
        )
        load_time = time.time() - start_load_time
        print(f"Model loaded successfully in {load_time:.2f} seconds.")

        print("\nAttempting to load tokenizer...")
        tokenizer = AutoTokenizer.from_pretrained(hf_model_name)
        print("Tokenizer loaded successfully.")

        # --- Define the translation function for a single text line ---
        def translate_line(text_line, model, tokenizer):
            if not text_line.strip(): # تخطي الأسطر الفارغة
                return ""
            try:
                inputs = tokenizer(text_line, return_tensors="pt", truncation=True) # Added truncation
                outputs = model.generate(**inputs)
                translated_text = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
                return translated_text
            except Exception as e:
                print(f"  Error translating line: '{text_line[:50]}...': {e}")
                return "[TRANSLATION ERROR]" # إرجاع علامة خطأ

        # --- Define the file translation function ---
        def translate_file(input_path, output_path, model, tokenizer):
            print(f"\nStarting translation of file: {input_path}")
            print(f"Output will be saved to: {output_path}")
            start_translate_time = time.time()
            lines_processed = 0
            lines_skipped = 0
            lines_error = 0

            try:
                with open(input_path, 'r', encoding='utf-8') as infile, \
                     open(output_path, 'w', encoding='utf-8') as outfile:

                    for line in infile:
                        original_line = line.strip() # إزالة المسافات البادئة واللاحقة
                        if not original_line:
                            outfile.write("\n") # كتابة سطر فارغ إذا كان السطر الأصلي فارغًا
                            lines_skipped += 1
                            continue

                        translated = translate_line(original_line, model, tokenizer)

                        if translated == "[TRANSLATION ERROR]":
                            lines_error += 1
                        elif not translated: # إذا كانت الترجمة فارغة لسبب ما
                             lines_skipped += 1

                        outfile.write(translated + '\n')
                        lines_processed += 1
                        if lines_processed % 50 == 0: # طباعة التقدم كل 50 سطرًا
                            print(f"  Processed {lines_processed} lines...")

            except FileNotFoundError:
                print(f"Error: Input file not found at {input_path}")
                return
            except Exception as e:
                print(f"An error occurred during file processing: {e}")
                return

            end_translate_time = time.time()
            total_time = end_translate_time - start_translate_time
            print("\n--- Translation Summary ---")
            print(f"File translation completed.")
            print(f"Total lines processed: {lines_processed}")
            if lines_skipped > 0: print(f"Lines skipped (empty or translation empty): {lines_skipped}")
            if lines_error > 0: print(f"Lines with translation errors: {lines_error}")
            print(f"Total time taken: {total_time:.2f} seconds.")
            print(f"Translated file saved to: {output_path}")

        # --- Execute the file translation ---
        if os.path.exists(input_file_path):
             translate_file(input_file_path, output_file_path, model, tokenizer)
        else:
             print(f"\nError: Input file '{input_file_path}' does not exist. Cannot start translation.")


    except FileNotFoundError as e:
        print(f"\nError loading model or tokenizer: {e}")
        print("This might indicate missing ONNX files (encoder/decoder/decoder_with_past) or config/tokenizer files.")
        print(f"Please ensure the necessary files exist in {onnx_model_path}.")
        if os.path.exists(onnx_model_path): print("Files found:", os.listdir(onnx_model_path)[:10])
    except Exception as e:
         print(f"\nAn unexpected error occurred during setup: {e}")
```

```python
import torch
from transformers import AutoTokenizer
from optimum.onnxruntime import ORTModelForSeq2SeqLM
import os
import shutil
import time

# --- Define Paths ---
parent_dir = "/content/opus-mt-en-ar"
onnx_model_path = os.path.join(parent_dir, "onnx")
hf_model_name = "Helsinki-NLP/opus-mt-en-ar"

# --- !!! قم بتحديد مسار الملف المدخل والمخرج هنا !!! ---
input_file_path = "/content/2.txt"      # استبدل هذا بمسار ملف الإدخال الفعلي
output_file_path = "/content/output_arabic_batched.txt" # استبدل هذا بالمسار المرغوب لملف الإخراج

# --- *** تحديد حجم الدفعة (Batch Size) *** ---
# يمكنك تجربة قيم مختلفة (مثل 16, 32, 64) للعثور على الأفضل لجهازك وحجم ذاكرتك
batch_size = 32

# --- (اختياري) إنشاء ملف إدخال نموذجي إذا لم يكن موجودًا ---
if not os.path.exists(input_file_path):
    print(f"ملف الإدخال '{input_file_path}' غير موجود. سيتم إنشاء ملف نموذجي.")
    # إنشاء ملف أكبر قليلاً كمثال للـ batching
    with open(input_file_path, 'w', encoding='utf-8') as f:
        for i in range(200): # إنشاء 200 سطر
             f.write(f"This is line number {i+1} for batch processing example.\n")
        f.write("\n") # سطر فارغ
        f.write("Another sentence here.\n")

# --- Ensure ONNX directory exists ---
os.makedirs(onnx_model_path, exist_ok=True)

# --- Copy necessary config/tokenizer files ---
print(f"Copying config/tokenizer files from {parent_dir} to {onnx_model_path}")
# (الكود الخاص بنسخ الملفات يبقى كما هو)
files_to_copy = [
    "config.json",
    "generation_config.json",
    "source.spm",
    "target.spm",
    "tokenizer_config.json",
    "vocab.json",
    "special_tokens_map.json"
]
copied_count = 0
for filename in files_to_copy:
    src_file = os.path.join(parent_dir, filename)
    dst_file = os.path.join(onnx_model_path, filename)
    if os.path.exists(src_file):
        if not os.path.exists(dst_file):
            try:
                shutil.copy2(src_file, dst_file)
                print(f"  Copied: {filename}")
                copied_count += 1
            except Exception as e:
                print(f"  Error copying {filename}: {e}")
        else:
             pass
    else:
        print(f"  Warning: Source file not found, skipping: {src_file}")

# --- Check if essential config file exists ---
config_present = os.path.exists(os.path.join(onnx_model_path, "config.json"))

if not config_present:
     print("\nError: Configuration file 'config.json' is missing in the ONNX directory.")
else:
    # --- Verification Step ---
    print(f"\nChecking files in: {onnx_model_path}")
    if os.path.exists(onnx_model_path):
        all_files = os.listdir(onnx_model_path)
        print(all_files[:10], "..." if len(all_files) > 10 else "")
    else:
        print("ONNX model directory not found!")

    # --- Load the ONNX Model and Tokenizer ---
    try:
        print(f"\nAttempting to load model (use_cache=True)...")
        start_load_time = time.time()
        model = ORTModelForSeq2SeqLM.from_pretrained(
            onnx_model_path,
            use_cache=True
        )
        load_time = time.time() - start_load_time
        print(f"Model loaded successfully in {load_time:.2f} seconds.")

        print("\nAttempting to load tokenizer...")
        tokenizer = AutoTokenizer.from_pretrained(hf_model_name)
        print("Tokenizer loaded successfully.")

        # --- Define the translation function for a BATCH of lines ---
        def translate_batch(lines_batch, model, tokenizer):
            if not lines_batch: # إذا كانت الدفعة فارغة
                return []
            try:
                # **مهم:** استخدام padding=True لتوحيد طول الأسطر في الدفعة
                inputs = tokenizer(lines_batch, return_tensors="pt", padding=True, truncation=True)
                outputs = model.generate(**inputs)
                translated_texts = tokenizer.batch_decode(outputs, skip_special_tokens=True)
                return translated_texts
            except Exception as e:
                print(f"  Error translating batch (first line: '{lines_batch[0][:50]}...'): {e}")
                # إرجاع قائمة بنفس حجم الدفعة ولكن مع علامات خطأ
                return ["[BATCH TRANSLATION ERROR]"] * len(lines_batch)

        # --- Define the file translation function using BATCHING ---
        def translate_file_batched(input_path, output_path, model, tokenizer, batch_sz):
            print(f"\nStarting batched translation of file: {input_path} (Batch size: {batch_sz})")
            print(f"Output will be saved to: {output_path}")
            start_translate_time = time.time()
            lines_processed = 0
            batches_processed = 0
            lines_in_batch = [] # لتخزين أسطر الدفعة الحالية

            try:
                with open(input_path, 'r', encoding='utf-8') as infile, \
                     open(output_path, 'w', encoding='utf-8') as outfile:

                    for line in infile:
                        original_line = line.strip()
                        if not original_line:
                            # التعامل مع الأسطر الفارغة: إذا كانت هناك دفعة قيد التجهيز، قم بترجمتها أولاً
                            if lines_in_batch:
                                translated_batch = translate_batch(lines_in_batch, model, tokenizer)
                                for translated_line in translated_batch:
                                    outfile.write(translated_line + '\n')
                                lines_processed += len(lines_in_batch)
                                batches_processed += 1
                                lines_in_batch = [] # إفراغ الدفعة
                                if batches_processed % 10 == 0: # طباعة التقدم كل 10 دفعات
                                     print(f"  Processed {batches_processed} batches ({lines_processed} lines)...")
                            # كتابة السطر الفارغ في المخرج
                            outfile.write('\n')
                            continue # الانتقال للسطر التالي

                        # إضافة السطر غير الفارغ للدفعة الحالية
                        lines_in_batch.append(original_line)

                        # إذا اكتملت الدفعة، قم بترجمتها وكتابتها
                        if len(lines_in_batch) >= batch_sz:
                            translated_batch = translate_batch(lines_in_batch, model, tokenizer)
                            for translated_line in translated_batch:
                                outfile.write(translated_line + '\n')
                            lines_processed += len(lines_in_batch)
                            batches_processed += 1
                            lines_in_batch = [] # إفراغ الدفعة للبدء من جديد
                            if batches_processed % 10 == 0: # طباعة التقدم كل 10 دفعات
                                 print(f"  Processed {batches_processed} batches ({lines_processed} lines)...")

                    # --- معالجة أي أسطر متبقية في الدفعة الأخيرة ---
                    if lines_in_batch:
                        print(f"  Processing final batch ({len(lines_in_batch)} lines)...")
                        translated_batch = translate_batch(lines_in_batch, model, tokenizer)
                        for translated_line in translated_batch:
                            outfile.write(translated_line + '\n')
                        lines_processed += len(lines_in_batch)
                        batches_processed += 1

            except FileNotFoundError:
                print(f"Error: Input file not found at {input_path}")
                return
            except Exception as e:
                print(f"An error occurred during file processing: {e}")
                return

            end_translate_time = time.time()
            total_time = end_translate_time - start_translate_time
            print("\n--- Translation Summary ---")
            print(f"File translation completed.")
            print(f"Total lines processed: {lines_processed}")
            print(f"Total batches processed: {batches_processed}")
            print(f"Total time taken: {total_time:.2f} seconds.")
            if lines_processed > 0 and total_time > 0:
                 print(f"Average lines per second: {lines_processed / total_time:.2f}")
            print(f"Translated file saved to: {output_path}")

        # --- Execute the file translation using batching ---
        if os.path.exists(input_file_path):
             translate_file_batched(input_file_path, output_file_path, model, tokenizer, batch_size)
        else:
             print(f"\nError: Input file '{input_file_path}' does not exist. Cannot start translation.")

    except FileNotFoundError as e:
        print(f"\nError loading model or tokenizer: {e}")
        # (رسائل الخطأ تبقى كما هي)
    except Exception as e:
         print(f"\nAn unexpected error occurred during setup: {e}")
```

```python
import torch
from transformers import AutoTokenizer, pipeline
from optimum.onnxruntime import ORTModelForSeq2SeqLM

# تحميل النموذج وال tokenizer
model = ORTModelForSeq2SeqLM.from_pretrained("opus_onnx")
tokenizer = AutoTokenizer.from_pretrained("opus_onnx")

# إنشاء خط أنابيب للترجمة
translator = pipeline(
    "translation",
    model=model,
    tokenizer=tokenizer,
    device="cuda" if torch.cuda.is_available() else "cpu"
)

# نص للترجمة (يمكنك تغييره حسب الحاجة)
input_text = "This is a sample text to translate."

# تنفيذ الترجمة
translated_result = translator(input_text)

# عرض النتيجة
print("النص الأصلي:", input_text)
print("النص المترجم:", translated_result[0]['translation_text'])
```
